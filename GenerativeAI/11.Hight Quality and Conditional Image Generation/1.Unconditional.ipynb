````python
from diffusers import DiffusionPipeline
import torch
import matplotlib.pyplot as plt
````

# Code

````python
def load_model(pretrained_model_or_path):
    pipeline = DiffusionPipeline.from_pretrained(pretrained_model_or_path)
    return pipeline
````

````python
def generate_image(pipeline, num_inference_steps=50):
    image = pipeline(num_inference_steps=num_inference_steps).images[0]
    return image
````

````python
def show_image(image):
    plt.axis('off')
    plt.imshow(image)
    plt.show()
````

````python
def main():
    pretrained_model_or_path = "CompVis/ldm-celebahq-256"
    num_inference_steps = 50
    model = load_model(pretrained_model_or_path)
    image = generate_image(model, num_inference_steps)
    show_image(image)
````

# Execution

````python
main()
````
