````python
import torch
from diffusers import  StableDiffusionPipeline
import matplotlib.pyplot as plt
````

image.png

# Code

````python
def load_model(model_id):
    pipeline = StableDiffusionPipeline.from_pretrained(model_id,  torch_dtype=torch.float16).to("cuda")
    return pipeline
````

````python
def generate_image(pipeline, prompt, num_inference_steps):
    result = pipeline(prompt=prompt, num_inference_steps=num_inference_steps)
    return result.images[0]
````

````python
def show_image(image):
    plt.axis('off')
    plt.imshow(image)
    plt.show()
````

````python
def main():
    model_id = "nousr/robo-diffusion-2-base"
    prompt = "A realistic image of a Robot."
    num_inference_steps = 32

    pipeline = load_model(model_id)
    image = generate_image(pipeline, prompt, num_inference_steps)
    show_image(image)
````

# Execution

````python
main()
````
